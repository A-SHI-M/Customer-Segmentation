```python
# Cell 1 — Imports & Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_theme(style='whitegrid')
plt.rcParams['figure.dpi'] = 100
pd.set_option('display.max_columns', None)
```

```python
# Cell 2 — Load Data
df = pd.read_csv('../data/ecommerce_customer_churn_dataset.csv')
print(f'Loaded {df.shape[0]:,} rows and {df.shape[1]} columns')
```

```python
# Cell 3 — Dataset Overview
print('--- Shape ---')
print(df.shape)

print('\n--- Data Types ---')
print(df.dtypes)

print('\n--- First 10 Rows ---')
df.head(10)
```

```python
# Cell 3b — Descriptive Statistics
df.describe(include='all').T
```

```python
# Cell 4 — Missing Values
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing %', ascending=False)

print('Columns with missing values:')
print(missing_df)

fig, ax = plt.subplots(figsize=(8, 4))
missing_df['Missing %'].plot(kind='bar', ax=ax, color='steelblue', edgecolor='black')
ax.set_title('Missing Values (% per Column)')
ax.set_ylabel('Missing %')
ax.set_xlabel('')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Cell 5 — Target Distribution (Churned)
churn_counts = df['Churned'].value_counts()
churn_pct = df['Churned'].value_counts(normalize=True) * 100
print('Churned value counts:')
print(churn_counts)
print('\nChurned %:')
print(churn_pct.round(2))

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].pie(
    churn_counts,
    labels=['Not Churned (0)', 'Churned (1)'],
    autopct='%1.1f%%',
    colors=['#4CAF50', '#F44336'],
    startangle=90
)
axes[0].set_title('Churn Distribution')

sns.countplot(x='Churned', data=df, palette=['#4CAF50', '#F44336'], ax=axes[1])
axes[1].set_title('Churn Count')
axes[1].set_xticklabels(['Not Churned', 'Churned'])
for p in axes[1].patches:
    axes[1].annotate(f'{int(p.get_height()):,}', (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Cell 6 — Numerical Distributions
num_cols = df.select_dtypes(include=np.number).columns.drop('Churned').tolist()
n_cols = 4
n_rows = int(np.ceil(len(num_cols) / n_cols))

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 3.5))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    axes[i].hist(df[col].dropna(), bins=40, color='steelblue', edgecolor='white', alpha=0.85)
    axes[i].set_title(col, fontsize=9)
    axes[i].set_xlabel('')

for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)

fig.suptitle('Numerical Feature Distributions', fontsize=14, y=1.01)
plt.tight_layout()
plt.show()
```

```python
# Cell 7 — Categorical Distributions
cat_cols = ['Gender', 'Country', 'Signup_Quarter']

fig, axes = plt.subplots(1, 3, figsize=(16, 5))
for ax, col in zip(axes, cat_cols):
    order = df[col].value_counts().index
    sns.countplot(y=col, data=df, order=order, palette='Blues_r', ax=ax)
    ax.set_title(f'{col} Distribution')
    ax.set_xlabel('Count')
    ax.set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
# Cell 8 — Correlation Heatmap
corr = df[num_cols + ['Churned']].corr()

fig, ax = plt.subplots(figsize=(16, 12))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(
    corr,
    mask=mask,
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    center=0,
    linewidths=0.5,
    ax=ax,
    annot_kws={'size': 7}
)
ax.set_title('Correlation Heatmap (Numerical Features + Churned)', fontsize=13)
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.yticks(fontsize=8)
plt.tight_layout()
plt.show()

print('\nTop features correlated with Churned:')
print(corr['Churned'].drop('Churned').abs().sort_values(ascending=False).head(10))
```

```python
# Cell 9 — Churn Rate by Category
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

for ax, col in zip(axes, ['Gender', 'Country', 'Signup_Quarter']):
    churn_rate = df.groupby(col)['Churned'].mean().sort_values(ascending=False) * 100
    churn_rate.plot(kind='bar', ax=ax, color='coral', edgecolor='black')
    ax.set_title(f'Churn Rate by {col}')
    ax.set_ylabel('Churn Rate (%)')
    ax.set_xlabel('')
    ax.set_ylim(0, churn_rate.max() * 1.2)
    plt.setp(ax.get_xticklabels(), rotation=30, ha='right')
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.1f}%',
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
# Cell 10 — Key Features vs Churn (Box Plots)
top_features = (
    corr["Churned"].drop("Churned").abs()
    .sort_values(ascending=False)
    .head(8)
    .index.tolist()
)

n_cols = 4
n_rows = 2
fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 9))
axes = axes.flatten()

for i, col in enumerate(top_features):
    sns.boxplot(
        x="Churned", y=col, data=df,
        hue="Churned",
        palette={0: "#4CAF50", 1: "#F44336"},
        legend=False,
        ax=axes[i]
    )
    axes[i].set_title(col, fontsize=9)
    axes[i].set_xticks([0, 1])
    axes[i].set_xticklabels(["Not Churned", "Churned"], fontsize=8)
    axes[i].set_xlabel("")

fig.suptitle("Top Features vs Churn", fontsize=13)
plt.tight_layout()
plt.show()
```

## Summary / Key Observations

- **Dataset**: 50,000 customers, 25 features, binary churn target.
- **Missing values**: A small number of columns have missing data (`Pages_Per_Session`, `Product_Reviews_Written`, `Social_Media_Engagement_Score`, `Payment_Method_Diversity`) — will require imputation in preprocessing.
- **Class balance**: Check the pie chart — if heavily imbalanced, resampling (SMOTE/undersampling) will be needed during modeling.
- **Correlations**: The heatmap and top-features box plots reveal which numerical features are most predictive of churn.
- **Categorical trends**: Churn rates vary across countries and gender — these features will be important for the model.
